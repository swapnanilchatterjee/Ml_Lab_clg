```python
import pandas as pd
import numpy as np
```

```python
df2=pd.read_csv("test_data - test_data (1).csv")
df1=pd.read_csv("train_data - train_data.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
x=df1[['x']]
y=df2[['y']]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn import linear_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
y_pred=clf.predict(x_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_test, y_test, color='red', label="Actual Data")  # Red points for actual data
plt.plot(x_test, y_pred, color='blue', linewidth=2, label="Predicted Line")  # Blue line for regression

# Labels and Title
plt.xlabel("X Test Data")
plt.ylabel("Y Values")
plt.title("Linear Regression Model - Test Data")

plt.legend()
plt.show()
```
